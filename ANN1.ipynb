```python
import numpy as np
import pandas as pd
import tensorflow as tf

dt= pd.read_csv('Churn_Modelling.csv')
X=dt.iloc[:, 3:-1].values
Y=dt.iloc[:, -1].values
#print(Y)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,2]=le.fit_transform(X[:,2])
#print(X)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct= ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder= 'passthrough')
X=np.array(ct.fit_transform(X))
#print(X)

from sklearn.model_selection import train_test_split
xt1,xt2,yt1,yt2= train_test_split(X, Y, test_size=0.2, random_state=0)

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
xt1= sc.fit_transform(xt1)
xt2= sc.fit_transform(xt2)
#print(xt1)

ann= tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=12, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann.fit(xt1, yt1, batch_size=32, epochs=100)

y_pred= ann.predict(xt2)
y_pred= (y_pred>0.5)

from sklearn.metrics import accuracy_score, confusion_matrix
cm= confusion_matrix(yt2, y_pred)
print(cm)
accuracy_score(yt2, y_pred)

#print(ann.predict(sc.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]]))>0.5)

```
