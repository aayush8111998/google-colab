```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset= pd.read_csv('TCS.NS.csv')
dataset.dropna(how= "all", inplace= True)
dataset.replace([np.inf, -np.inf], np.nan, inplace=True)

dataset= dataset.drop(dataset.index[[40, 1185]])


X= dataset.iloc[:, ~dataset.columns.isin(['Close','Date'])].values
Z= dataset.iloc[:, 0].values
Y= dataset.iloc[:, 3].values
#print(X)



from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
Z= le.fit_transform(Z)


#print(Z)



from sklearn.ensemble import RandomForestRegressor
reg= RandomForestRegressor(n_estimators=5, random_state=0)
reg.fit(X,Y,Z)


reg.predict([[1000,1000,1000,1000,1000]])



plt.plot(Z, reg.predict(X), color='blue')
plt.show()








```
